g[L1,1] at 1s:
No gradients before this, so staleness = 0.

g[L1,2] at 2s:
Since no other learner updated in between, staleness = 0.

g[L1,3] at 3s:
Only g[L2,1] happens at 2.5s after g[L1,2], but after L1's last pull at 2s. So,

Between L1's 2nd and 3rd update, g[L2,1] happens at 2.5s.

So, staleness = 1 (because L2 updated the weights once since L1 last pulled).

g[L1,4] at 4s:
Since L2’s next update (g[L2,2]) is at 5s (after 4s), no updates between L1's 3rd and 4th, so staleness = 0.

g[L2,1] at 2.5s:
L2's last pull would have been after L2’s previous update (none in this window), so since L1 updated at 1s and 2s before 2.5s, number of L1 updates since last L2 pull = 2.

So, staleness = 2.

g[L2,2] at 5s:
Since L2’s last pull (2.5s), L1 did updates at 3s and 4s.

So, staleness = 2.
