## Q1

From each input vector, three vectors are derived in a self-attention block:

The Query vector, the Key vector, the value vector.

Each input is projected into these three vectors using learned weight matrices.



## Q2

For each pair of input positions, we compute the dot product of the query of one position and the key of another position.

This gives us a score indicating how much attention the model should pay from one position to another.

These scores are divided by the square root of the dimension of the key vectors (for numerical stability), and then a softmax is applied across all positions to obtain the attention weights.

These weights are used to take a weighted sum of the value vectors, producing the output for each position.

## Q3

Each head requires three weight matrices: one each for Q, K, and V.

For 8 heads: 8 (heads) × 3 (matrices) = 24 weight matrices (for Q, K, V projections).

Additionally, there is one more weight matrix for the output projection, which combines all heads back into a single output.

Total weight matrices = 8 × 3 + 1 = 25

## Q4

The output from each attention head is concatenated (not added), resulting in a large vector (if each head outputs size 64 and there are 8 heads, you get a vector of size 8 × 64 = 512).

This concatenated vector is then projected back to the original dimension (e.g., 512) using a learned linear transformation (output weight matrix).

This projection produces a single vector per word, which is the expected input format for the following feed-forward layer.
